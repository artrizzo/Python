```python
import networkx as nx

cor_vertices = [nx.degree_centrality(projeto)[v] for v in projeto]
tamanho_vertices = [300 + 300 * projeto.degree(v) for v in projeto]

rotulo_vertices = {} 
for v in projeto.nodes(): 
  rotulo_vertices[v] = round(nx.degree_centrality(projeto)[v], 2)

pos = nx.spring_layout(projeto)
print("Centralidade de Grau: ")
plt.figure(figsize = (20, 20))
nx.draw(projeto, pos, node_size = tamanho_vertices, node_color = cor_vertices)
nx.draw_networkx_labels(projeto, pos, rotulo_vertices, font_weight = 'bold', font_color = 'white' );
```

```python
import networkx as nx

cor_vertices = [nx.closeness_centrality(projeto)[v] for v in projeto]
tamanho_vertices = [250 + 250 * projeto.degree(v) for v in projeto]

rotulo_vertices = {} 
for v in projeto.nodes(): 
  rotulo_vertices[v] = round(nx.closeness_centrality(projeto)[v], 2)

print("Centralidade de Proximidade:")
plt.figure(figsize = (20, 20))
nx.draw(projeto, pos, node_size = tamanho_vertices, node_color = cor_vertices)
nx.draw_networkx_labels(projeto, pos, rotulo_vertices, font_weight = 'bold', font_color = 'white' );
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# centralidade relativa de intermediação

cor_vertices = [nx.betweenness_centrality(projeto)[v] for v in projeto]
tamanho_vertices = [260 + 260 * projeto.degree(v) for v in projeto]

rotulo_vertices = {} 
for v in projeto.nodes(): 
  rotulo_vertices[v] = round(nx.betweenness_centrality(projeto)[v], 2)

plt.figure(figsize =(12, 12))
print("Centralidade de intermediação:")
nx.draw(projeto, pos, node_size = tamanho_vertices, node_color = cor_vertices)
nx.draw_networkx_labels(projeto, pos, rotulo_vertices, font_weight = 'bold', font_color = 'white' );
```
