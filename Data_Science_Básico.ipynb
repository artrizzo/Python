#def e funções

```python
def funçaotexto(texto):
  return texto[0]+texto[-1]+' - pegar a primeira e a última letra'
```

```python
texto = 'Arthur Rizzo'
```

```python
funçaotexto(texto)
```

```python
def funçaotexto2(texto = 'Arthur Rizzo'):
  return texto[0]+texto[-1]+' - pegar a primeira e a última letra'
```

```python
funçaotexto2()
```

Lambda - é um def para funções mais simples, sem muitas variáveis
